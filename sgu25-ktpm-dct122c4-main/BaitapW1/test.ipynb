```python
# Test cases với validation
test_cases = [
    ("0 0 0", "Vô số nghiệm", "Infinite solutions"),
    ("0 0 5", "Vô nghiệm (a=b=0, c≠0)", "No solution"),
    ("0 1 -4", "a=0, có nghiệm", "2 real solution(s): 2 -2"),
    ("0 1 4", "a=0, vô nghiệm", "No solution"),
    ("1 0 1", "Delta < 0", "No solution"),
    ("1 -2 1", "Delta = 0", "2 real solution(s): 1 -1"),
    ("1 -5 4", "4 nghiệm thực", "4 real solution(s)"),
    ("1 0 -1", "2 nghiệm thực", "2 real solution(s): 1 -1"),
]

for i, (input_data, description, expected) in enumerate(test_cases, 1):
    print(f"\n=== Test Case {i}: {description} ===")
    print(f"Input: {input_data}")
    
    # Chạy chương trình
    result = subprocess.run(
        ["./phuongtrinhtrungphuong.exe"], 
        input=input_data, 
        capture_output=True, 
        text=True
    )
    
    actual = result.stdout.strip()
    print(f"Expected: {expected}")
    print(f"Actual: {actual}")
    
    # Kiểm tra
    if expected in actual:
        print(" PASS")
    else:
        print(" FAIL")
    print("-" * 50)
```
